# Renombrado y Combinación

En este JNB se revisará el tema de Renombrado y Combinación en Pandas.

```python
import pandas as pd
pd.set_option('display.max_rows', 5)
reviews = pd.read_csv("https://raw.githubusercontent.com/GTD2101-ACD-GB-EJ25/ACD.datasets.kaggle/refs/heads/main/winemag-data-130k-v2.txt", index_col=0)
```

```python
reviews.head()
```

```python
reviews.rename(columns={'points': 'score'})
```

```python
reviews.rename(index={0: 'firstEntry', 1: 'secondEntry'})
```

```python
reviews.rename(index={0: 'firstEntry', 1: 'secondEntry'}).loc['firstEntry']
```

```python
reviews.rename_axis("wines", axis='rows').rename_axis("fields", axis='columns')
```

```python
canadian_youtube = pd.read_csv("https://raw.githubusercontent.com/GTD2101-ACD-GB-EJ25/ACD.datasets.kaggle/refs/heads/main/CAvideos.txt")
british_youtube = pd.read_csv("https://raw.githubusercontent.com/GTD2101-ACD-GB-EJ25/ACD.datasets.kaggle/refs/heads/main/GBvideos.txt")

pd.concat([canadian_youtube, british_youtube])
```

```python
canadian_youtube
```

```python
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix='_CAN', rsuffix='_UK')
```
